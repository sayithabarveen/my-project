```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
print("Food Delivery Sales & Customer Insights")
```

```python
# TASK 1: Analyze order data - time, location, item
from google.colab import files
import pandas as pd
uploaded = files.upload()
df = pd.read_csv('food_delivery_orders.csv')

print("=== DATASET PREVIEW ===")
display(df.head())

print("\n=== DATASET INFO ===")
df.info()

print("\n=== MISSING VALUES CHECK ===")
print(df.isnull().sum())
```

```python
# DATA CLEANING - Required for accurate analysis
df = df.drop_duplicates()
df['Price'].fillna(df['Price'].median(), inplace=True)
df['Order_Time'] = pd.to_datetime(df['Order_Time'])
df['Order_Hour'] = df['Order_Time'].dt.hour
df['Order_Day'] = df['Order_Time'].dt.day_name()

print("=== CLEANED DATA ===")
display(df.head())
print("\nData cleaning completed. Ready for analysis.")
```

```python
# TASK 2: Identify peak order times
import matplotlib.pyplot as plt
hourly_orders = df.groupby('Order_Hour')['Order_ID'].count().reset_index()
hourly_orders.columns = ['Hour', 'Total_Orders']

plt.bar(hourly_orders['Hour'], hourly_orders['Total_Orders'], color='#FF6B35')
plt.title('Task 2: Peak Order Times Analysis', fontsize=16, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.xticks(range(0,24))
for i, v in enumerate(hourly_orders['Total_Orders']):
    plt.text(hourly_orders['Hour'][i], v + 0.05, str(v), ha='center')
plt.show()

peak_hour = hourly_orders.loc[hourly_orders['Total_Orders'].idxmax(), 'Hour']
print(f"INSIGHT: Peak ordering time is {peak_hour}:00 - {peak_hour+1}:00")
```

```python
# TASK 3: Find most popular dishes
import seaborn as sns
top_dishes = df['Dish_Name'].value_counts().head(5)

plt.pie(top_dishes.values, labels=top_dishes.index, autopct='%1.1f%%',
        startangle=90, colors=sns.color_palette('Set2'))
plt.title('Task 3: Most Popular Dishes', fontsize=16, fontweight='bold')
plt.show()

print(f"INSIGHT: Most popular dish is {top_dishes.index[0]} with {top_dishes.values[0]} orders")
```

```python
# TASK 4: Create a dashboard for business insights
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Food Delivery Sales & Customer Insights Dashboard', fontsize=20, fontweight='bold')

# Chart 1: Orders by Location
location_orders = df['Delivery_Location'].value_counts()
axes[0,0].barh(location_orders.index, location_orders.values, color='skyblue')
axes[0,0].set_title('Orders by Delivery Location')
axes[0,0].set_xlabel('Number of Orders')

# Chart 2: Price Distribution
axes[0,1].hist(df['Price'], bins=5, color='lightgreen', edgecolor='black')
axes[0,1].set_title('Order Price Distribution')
axes[0,1].set_xlabel('Price in Rs')
axes[0,1].set_ylabel('Frequency')

# Chart 3: Day-wise Orders
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
daily_orders = df['Order_Day'].value_counts().reindex(day_order).fillna(0)
axes[1,0].plot(daily_orders.index, daily_orders.values, marker='o', color='red', linewidth=2)
axes[1,0].set_title('Orders by Day of Week')
axes[1,0].set_ylabel('Number of Orders')
axes[1,0].tick_params(axis='x', rotation=45)

# Chart 4: Customer Rating Analysis
axes[1,1].boxplot(df['Customer_Rating'], vert=False)
axes[1,1].set_title('Customer Rating Analysis')
axes[1,1].set_xlabel('Rating out of 5')

plt.tight_layout()
plt.show()
```

```python
# FINAL INSIGHTS FOR TEMPLATE
print("="*50)
print("BUSINESS INSIGHTS SUMMARY - FOR REPORT")
print("="*50)
print(f"1. Peak Order Time: {df['Order_Hour'].mode()[0]}:00 hours")
print(f"2. Most Popular Dish: {df['Dish_Name'].mode()[0]}")
print(f"3. Busiest Location: {df['Delivery_Location'].mode()[0]}")
print(f"4. Average Order Value: Rs. {round(df['Price'].mean(),2)}")
print(f"5. Average Customer Rating: {round(df['Customer_Rating'].mean(),1)}/5.0")
print("="*50)
print("Recommendation: Increase staff and stock during peak hours for top dishes.")
```
